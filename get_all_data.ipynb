```python
grouped_data = data.groupby(['Notetype', 'NoteTemplate']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
grouped_data.head(30)
```

```python
grouped_data.to_csv('grouped_data.csv', index=False)
```

```python
# Calculate total counts
total_counts = grouped_data['counts'].sum()

# Calculate percentages using vectorized operations
grouped_data['Percentage'] = (grouped_data['counts'] / total_counts) * 100

```

```python
# Calculate cumulative percentage
grouped_data['Cumulative Percentage'] = grouped_data['Percentage'].cumsum()
```

```python
# Export the DataFrame
grouped_data.to_csv('grouped_data_with_percentages.csv', index=False)
```

```python
import pandas as pd
from IPython.display import display, HTML
import numpy as np
```

```python


# I'm assuming 'data' is your main DataFrame and 'grouped_data' contains the groups with counts.
# Loop over the first 30 groups
for index, row in grouped_data.head(30).iterrows():
    # Extract the Notetype and NoteTemplate
    notetype = row['Notetype']
    notetemplate = row['NoteTemplate']
    
    # Filter data to get entries matching the Notetype and NoteTemplate
    group_data = data[(data['Notetype'] == notetype) & (data['NoteTemplate'] == notetemplate)]
    
    # Check if there are at least 5 notes in the group
    if len(group_data) >= 5:
        # Select 5 random samples from this group
        samples = group_data.sample(5)
    else:
        # If less than 5, take all available
        samples = group_data
    
    # Print out the Notetype and NoteTemplate clearly
    display(HTML(f"<h2 style='color: blue; background-color: lightyellow; padding: 10px; border-radius: 5px;'>Displaying notes for Notetype: {notetype}, NoteTemplate: {notetemplate}</h2>"))
    
    # Display the selected SOAPTEXT from the samples
    for _, sample in samples.iterrows():
        display(HTML(sample['SOAPTEXT']))
```
